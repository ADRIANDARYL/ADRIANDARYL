```python
#Question c
import nltk
from nltk import CFG

grammar_c = CFG.fromstring("""
S -> A
S -> B

A -> 'a' 'a' A 'b'
A -> C

C -> 'a' C
C -> 'a'

B -> 'a' 'a' B 'b'
B -> D

D -> 'b' 'b' E
E -> 'b' E
E ->
""")

tests = [
    "a a a b",
    "a a b",
    "a a a a b",
    "b b",
    "a a b b"
]

parser = nltk.RecursiveDescentParser(grammar_c)

for test in tests:
    print("Testing:", test)
    found = False

    for tree in parser.parse(test.split()):
        found = True
        print(tree)

    print("Accepted\n" if found else "Rejected\n")
```

```python
#Question d
import nltk
from nltk import CFG

grammar_d = CFG.fromstring("""
S -> 'a' S 'b' 'b'
S -> 'a' S 'b' 'b' 'b'
S ->
""")

tests = [
    "",
    "a b b",
    "a b b b",
    "a a b b b b",
    "a b"
]

parser = nltk.RecursiveDescentParser(grammar_d)

for test in tests:
    print("Testing:", repr(test))
    found = False

    for tree in parser.parse(test.split()):
        found = True
        print(tree)

    print("Accepted\n" if found else "Rejected\n")
```

```python
#Question e
import nltk
from nltk import CFG

grammar_e = CFG.fromstring("""
S -> A
S -> B

A -> E 'a' E
A -> E 'a' A

B -> E 'b' E
B -> E 'b' B

E -> 'a' E 'b' E
E -> 'b' E 'a' E
E -> E E
E ->
""")

tests = [
    "a",
    "b",
    "a a b",
    "a b",
    "a a b b",
    "a a a a b b"
]

parser = nltk.ChartParser(grammar_e)

for test in tests:
    print("Testing:", test)
    trees = list(parser.parse(test.split()))

    if trees:
        print("Accepted\n")
    else:
        print("Rejected\n")
```
